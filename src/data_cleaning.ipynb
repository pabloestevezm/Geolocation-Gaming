# Para comenzar con la búsqueda voy a filtrar para encontrar las startups con 'Tech" en su descripción y con más de 1M de captación

```python
tech_desc = re.compile("tech",re.IGNORECASE)
query = {
    "$and":[
        {"description":tech_desc},
        {"founded_year":{"$gte":2008}}
    ]
}

stcomp = db.companies.find(query,{"description":1,"name":1,"offices":1,"total_money_raised":1})
```

```python
starts= pd.DataFrame(stcomp)
starts
```

```python
starts['total_money_raised'] = starts['total_money_raised'].apply(total_funds)
#starts
stsucc = starts[starts.total_money_raised >= 1000000]
startupsoff = starts.explode("offices").apply(lambda e: e.offices,axis=1,result_type="expand")
#startupsoff['city'].value_counts().head(20)
stsucc
#startupsoff
```

```python
startupsoff.rename(columns={'description':'sec_description'},inplace=True)
```

```python
startups_offices = pd.concat([stsucc[["_id","name","description","total_money_raised"]], startupsoff],axis=1)
startups_offices
```

```python
startups_offices["geopoint"] = startups_offices.apply(transformToGeoPoint, axis=1)
startups_offices
```

```python
#Esto es para la limpieza de las filas que no voy a necesitar 
geostart = startups_offices[startups_offices['_id'].notna()]
#geostart
geostartups = geostart[geostart['geopoint'].notna()]
geostartups
```

```python
geostartups['city'].value_counts().head(20)
```

```python
ny_startups_offices = geostartups[geostartups.state_code == "NY"]
```

```python
ny_startups_offices
```

```python
ny_startups_offices["_id"] = ny_startups_offices["_id"].apply(lambda e: str(e))
```

```python
ny_startups_offices.to_json("Output/ny_startups_offices.json",orient="records")
```

```python
ny_startups_offices.to_csv("../Output/ny_startups_data.csv")
```

# También voy a filtrar para encontrar las empresas que se dediquen a design

```python
designcomp = re.compile("design",re.IGNORECASE)
query = {
    "description": designcomp
}
designcomp = db.companies.find(query,{"description":1,"name":1,"offices":1})
```

```python
designoff = pd.DataFrame(designcomp)
designoff
design_offs = designoff.explode("offices").apply(lambda e: e.offices,axis=1,result_type="expand")
design_offs
```

```python
design_offs.rename(columns={'description':'sec_description'},inplace=True)
```

```python
design_offices = pd.concat([designoff[["_id","name","description"]], design_offs],axis=1)
design_offices
```

```python
design_offices["geopoint"] = design_offices.apply(transformToGeoPoint, axis=1)
```

```python
design_offices
```

```python
geodesigns = design_offices[design_offices['geopoint'].notna()]
geodesigns
```

```python
geodesigns['city'].value_counts().head(20)
```

```python
ny_design_offices = geodesigns[geodesigns.city == "New York"]
ny_design_offices
```

```python
ny_design_offices_clean = ny_design_offices[['_id', 'name', 'description', 'address1', 'address2', 'latitude', 'longitude', 'geopoint']]
ny_design_offices_clean["_id"] = ny_design_offices_clean["_id"].apply(lambda e: str(e))
```

```python
ny_design_offices_clean.to_json("Output/ny_design_offices.json",orient="records")
```

```python
ny_design_offices.to_csv("../Output/ny_design_data.csv")
```
